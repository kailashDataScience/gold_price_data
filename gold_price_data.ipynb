```python
import pandas as pd 
import requests as req
from bs4 import BeautifulSoup
```

```python
url=["https://www.livechennai.com/get_goldrate_history.asp?monthno=1&yearno=2022",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=2&yearno=2022",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=3&yearno=2022",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=4&yearno=2022",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=5&yearno=2022",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=5&yearno=2022",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=7&yearno=2022",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=8&yearno=2022",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=9&yearno=2022",
      "https://www.livechennai.com/get_goldrate_history.asp?monthno=10&yearno=2022",
       "https://www.livechennai.com/get_goldrate_history.asp?monthno=11&yearno=2022",
       "https://www.livechennai.com/get_goldrate_history.asp?monthno=12&yearno=2022"

       
]
pages=[]
for page in url:
  pages.append(req.get(page))
```

```python

soups=[]
for page in pages:
  soups.append(BeautifulSoup(page.text,'lxml'))
```

```python

tr_tags=[]
for soup in soups:
  tr_tags.append(soup.find_all('tr'))

```

```python

def get_monthly_data(trlist):
    month_data=[]
    for month in trlist:
      for tr in month:
        td_tag=tr.find_all('td')
        row = [i.text.strip() for i in td_tag]
        month_data.append(row)
    return month_data
    
```

```python

monthly_data = get_monthly_data(tr_tags)
```

```python
string='Date'
for lst in monthly_data:
  length=len(lst)
  if length < 3 or string in lst:
    monthly_data.remove(lst)
monthly_data
```

```python
def process(data):
  return [lst[0] for lst in data],[lst[1] for lst in data],[lst[2] for lst in data]
```

```python
date,gold_24k,gold_22k=process(monthly_data)
```

```python
gold_data_dict={
      "DATE":date,
    "GOLD_24k_PRICE":gold_24k,
    "GOLD_22k_PRICE":gold_22k
  }
```

```python
df =pd.DataFrame(gold_data_dict)
```
