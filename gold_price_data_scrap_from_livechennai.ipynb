```python
import pandas as pd 
import requests as req
from bs4 import BeautifulSoup
```

```python
url="https://www.livechennai.com/get_goldrate_history.asp?monthno=1&yearno=2023"
page = req.get(url)
```

```python
soup = BeautifulSoup(page.text,'lxml')
```

```python
tr_tag=soup.find_all('tr')
data=[]
for tr in tr_tag:

  td_tag =tr.find_all('td')
  row = [i.text.strip() for i in td_tag]
  data.append(row)
```

```python
data = data[3:][:-3]
```

```python
Date      =    [lst[0] for lst in data]
gold_24k  =    [lst[1] for lst in data]
gold_22k  =    [lst[2] for lst in data]
```

```python
gold_price_data={
    "DATE":Date,
    "GOLD_24k_PRICE":gold_24k,
    "GOLD_22k_PRICE":gold_22k
}
```

```python
df =pd.DataFrame(gold_price_data)
```
